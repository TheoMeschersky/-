## ЧАСТЬ 3: создание и преобразование столбцов

#### Создайте колонку 'Top_University' со значением True для университетов, входящих в топ-50. Обратите внимание на значения типа "231–240".

```python
def Top(value):
    if pd.isna(value) or '–' in value:
        return False
    return int(value) <= 50

table['Top_University'] = table['University Global Ranking'].apply(Top)
table.sample(10) #Для проверки
```

#### Создайте колонку Academic_Performance, в которую внесите значение:
#### - "High", если GPA выше 3.5 баллов,
#### - "Medium", если GPA от 3 до 3.49 баллов,
#### - "Low", если GPA ниже 3 баллов.

```python
def perf(value):
  if pd.isna(value) or (3 <= float(value) <= 3.49):
    return 'Medium' #Будем считать, что у тех, у кого не известен GPA, он средний
  elif float(value) >= 3.5:
    return 'High'
  else:
    return 'Low'

# В строке 22 — мусор. Удалим её:
if 22 in table.index:  # чтобы удалить строчку только один раз и не получать ошибку
    table = table.drop(22, axis=0)

table['Academic_Performance'] = table['GPA (or Equivalent)'].apply(perf)
table.sample(10)
```

#### Посмотрите на уникальные значения в колонке Degree и приведите их к единообразию (Bachelor of Science -> Bachelor).

```python
#table['Degree'].unique()
'''Принятые замены:
MBA = Master of Business Administration
Bachelor = Bachelor of Science
Bachelor = Bachelor's Degree
Not stated = NaN
Dropout = Dropped Out
PhD = Ph.D.
Master = MD'''

def Degr(D):
  if pd.isna(D):
    return 'Not stated'
  elif D == 'Master of Business Administration':
    return 'MBA'
  elif D == "Bachelor of Science" or D == "Bachelor's Degree":
    return 'Bachelor'
  elif D == 'Dropped Out':
    return 'Dropout'
  elif D == 'Ph.D.':
    return 'PhD'
  elif D == 'MD':
    return 'Master'
  else:
    return D

table['Degree'] = table['Degree'].apply(Degr)
table.sample(10)
```
